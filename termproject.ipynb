## Term Project Proposal


#### 소프트웨어융합학과 2018103258 임형준


### 1. 주제

KBO 프로야구 선수들의 성적을 토대로 게임 내에서 합당한 선수카드의 능력치 설정


### 2. 가설
- 게임을 즐기는 많은 유저들이 게임 내 선수카드의 능력치가 합당하지 않다고 느끼는 경우가 많다.
- 야구 게임 "마구마구" 의 경우 한 시즌간 선수들의 능력치를 제대로 반영하여 선수카드 능력치를 측정하지 않은 경우가 존재 할 것이다.
- 인기팀 (인기 선수)의 경우 선수카드에게 좋은 능력치를 부여하고 비 인기팀 (비 인기 선수)의 경우 선수카드에 좋지 않은 능력치를 부여하는 경우가 존재할 것이다.


### 3. 분석할 데이터의 출처와 데이터 내용
- 온라인 야구 게임 마구마구 (https://ma9.netmarble.net/main.asp) 내 능력치를 가져와서 분석 할 예정이다.
- "마구마구" 의 경우
    * 투수 능력치는 체력, 제구력, 구속, 각 투수만의 구종 능력치 가 존재한다.
    * 타자 능력치는 파워, 컨택, 스피드, 스로잉, 수비 가 존재한다.
- 투수들의 능력치 측정에 있어서는 선수들의 구종에 대한 정보가 필요하지만 해당 데이터의 수집이 스탯티즈에서 어려운 관계로 **타자의 능력치** 만 측정을 할 예정이다.




- 스탯티즈 (http://www.statiz.co.kr/main.php) 홈페이지를 통해 선수들의 능력치를 가져와서 분석 할 예정이다.
- 타자의 경우 
    * "타격 일반 기록" 으로 타석, 타수, 득점, 안타, 2루타, 3루타, 홈런, 총 루타, 타점, 도루, 도루실패, 볼넷, 사구, 고의사구, 삼진, 병살, 희생번트, 희생플라이 존재한다.
    * "타격 비율 기록"으로는 타율, 출루율, 장타율, OPS, wOBA, wRC+, WAR, WPA가 존재한다.
    * "수비 일반 기록" 으로는 출장, 선발, 이닝, 기회, 자살, 보살, 실책이 존재한다. 
    * "수비 비율 기록" 으로는 수비율, sFR(RNG, ARM 등)이 존재한다.



### 4. 분석 계획


- 마구마구 게임 내 타자 능력치에 영향을 미칠만한 실제 기록 분류 (예시)
    * 파워 : 홈런, 장타율
    * 컨택 : 타율, 출루율, wOBA(타석당 득점 기여도), wRC+(득점 생산력)
    * 스피드 : 도루, 병살
    * 스로잉 : 보살, 외야수의 경우 ARM(보살율), 내야수의 경우 병살처리, 내야안타 허용 횟수
    * 수비 : 수비율, RNG(수비범위)

- 분석 계획
    * 위 처럼 관련이 있다고 생각하는 기록들을 모아 각 기록들에게 가중치 부여
    * 해당 가중치들을 모두 더한 후 평균과 최고, 최하점을 계산
    * 마구마구가 가진 능력치 분류를 통해 실제 기록의 가중치의 합을 기준으로 삼아 능력치 부여하기
- cf) 마구마구 능력치 분류 : 마구마구의 게임 능력치는 60~70 은 하, 71~80 은 중, 81~90은 상, 91~100은 최상, 101~ 110극상, 111~ 초월 로 분류된다.
- 게임 내 능력치와 실제 선수들의 기록을 분석하여 각 능력치에 적합한 가중치를 부과하여 해당 선수의 기록을 넣으면 적합한 능력치가 만들어지는 프로그램을 만들어볼 예정이다.
